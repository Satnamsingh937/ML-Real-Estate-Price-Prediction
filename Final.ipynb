```python
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel(r"C:\Users\ssatn\Desktop\Classeur1.xlsx")


df=df.loc[(df["Adresse"].str.contains("92130"))|(df["Adresse"].str.contains("92130"))]



def find_word(sentence, wordToFind):
    status = sentence.find(wordToFind)
    if status == -1 : 
        return 0
    else:
        return 1


df['Appartement'] = 0
df['Maison'] = 0
df['Studio'] = 0
df['Duplex'] = 0
for i in range(len(df)):
    text = df['Type'].iloc[i].lower()
    if find_word(text, 'appartement') == 1:
        df['Appartement'].iloc[i] = 1
    elif find_word(text, 'maison') == 1:
        df['Maison'].iloc[i] = 1
    elif find_word(text, 'studio') == 1:
        df['Studio'].iloc[i] = 1
    elif find_word(text, 'duplex') == 1:
        df['Duplex'].iloc[i] = 1

def Str_To_Float(x) : 
    if pd.isnull(x) :
        return np.nan
    elif x=="Non renseigné":
        return np.nan
    else : 
        x=x.split(" ")[0]
        x=x.replace(",",".")
        return x


def Str_To_Int(x) :
    if pd.isnull(x) :
        return np.nan
    elif x=="Non renseigné" :
        return np.nan
    else : 
        x=str(x)
        x=x.split(" ")[0]
        x=x.strip()
        x=int(x)
        return x
    
    
df['Nombres de Chambres']=df['Nombres de Chambres'].apply(Str_To_Int)



df["Etage"]=df["Etage"].fillna("Non Renseigné")




def Get_Floor(x) :
    if pd.isnull(x) :
        return np.nan
    else : 
        try : 
            if x == "Non renseigné" :
                return np.nan
            else : 
                return int(x.split()[1][0])
        except : 
            return np.nan


df["Etage"]=df["Etage"].apply(Get_Floor)



def Get_Price(x) :
    if pd.isnull(x) :
        return np.nan
    elif x=="Non renseigné" :
        return np.nan
    else : 
        x=str(x)
        x=x.strip()
        x=x.replace(" ","")
        x=x[:-1]
        x=int(x)
        return x
    
df["Prix"]=df["Prix"].apply(Get_Price)



        
df["Nombre de Pièces"]=df["Nombre de Pièces"].apply(Str_To_Int)

def Get_Surface(x) :
    if pd.isnull(x)  :
        return np.nan
    else : 
        try : 
            x=x.replace(",",".")
            x=x.split()[0]
            if x == "Non" : 
                return np.nan
            else : 
                x=float(x)
                return x
        except : 
            return np.nan
        
        
        
        
        
New=[]
for i in df["Surface"] : 
    if 'tage' in i : 
        i="Non renseigné"
        New.append(i)
    else :
        i=i
        New.append(i)
df["Surface"]=New

df['Surface']=df['Surface'].apply(Get_Surface)



df['Adresse']=df['Adresse'].str.lower()

def Splitage_Adresse_Quartier(x) :
    if pd.isnull(x) :
        return None
    else :
        x=x.lower()
        x=x.split("à")[0]
        return x.title()
    
    
def Splitage_Adresse_Ville(x) :
    if pd.isnull(x) :
        return None
    else :
        x=x.lower()
        x=x.split("à")[1]
        return x.title().replace("(","").replace(")","")
    
df["Quartier"]=df['Adresse'].apply(Splitage_Adresse_Quartier)
df.drop(columns='Adresse',inplace=True)




import numpy as np
import pandas as pd



critere=["cuisine","cuisine équipée","cuisine américaine équipée","ascenseur","Digicode","Interphone","Placards","Terrasse","Balcon","Salle de bain (baignoire)","Sans vis-à-vis","parking","cave","Jardin","Gardien","Chauffage individuel électrique","Chauffage individuel","Chauffage individuel au gaz","Chauffage collectif au gaz","exposition sud"]
critere = list(i.lower().strip() for i in critere)

df["Caractéristique1"]=df["Caractéristique1"].fillna("Non Renseigné")
df["Caractéristique2"]=df["Caractéristique2"].fillna("Non Renseigné")

C1=list(i.lower() for i in df["Caractéristique1"])
df["Caractéristique1"]=C1

C2=list(i.lower() for i in df["Caractéristique2"])
df["Caractéristique2"]=C2


def Type(x,y) :
    x=x[x.find(y)+len(y):].strip()[0]
    if not x in ("a","b","c","d","e","f") :
        return "Non Renseigné"
    else : 
        return x


Indice=[]
for i in list(df["Caractéristique2"]) : 
    Current_Indice=Type(i,"(dpe)")
    Indice.append(Current_Indice)
df["performance énergétique"]=Indice


df["Caractéristique3"]=df["Caractéristique1"].str[:]+" "+df["Caractéristique2"].str[:]
for i in critere :
    dlist=[]
    for j in list(df["Caractéristique3"]) :
        if i in j : 
            dlist.append(1)
        else : 
            dlist.append(0)
    df[i]=dlist
    
    
    
df.drop(columns=["Caractéristique1","Caractéristique2","Type"],inplace=True)



len(df['Quartier'].unique())



plt.bar(['Appartement', 'Maison', 'Studio','Duplex'], [df['Appartement'].sum(), df['Maison'].sum(), df['Studio'].sum(), df['Duplex'].sum()], color =['#045851','#24AFC8','#E1B81F','#E11F3C'], width = 0.5)
plt.xticks(fontweight="bold")
plt.yticks(fontweight="bold")
plt.title("Type", fontweight="bold", x=0.5, y=1.05)



from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit_transform(df['Quartier'])

df['Quartier'] = le.fit_transform(df['Quartier'])


df['Quartier'].unique()


df1 = df.dropna()

df1.info()

df1.drop(columns=["Caractéristique3"],inplace=True)

df1['performance énergétique'] = le.fit_transform(df1['performance énergétique'])


df1.info()

y = df1.iloc[:,4].values
df1.drop(columns = ['Prix','cuisine'],inplace=True)


X = df1.values


len(df1['Quartier'].unique())


from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(X , y , test_size = 0.20,random_state =2)


x_train[0]


from sklearn.ensemble import RandomForestRegressor


model1 = RandomForestRegressor()

model1.fit(x_train,y_train )


y_pred = model1.predict(x_test)


from sklearn.metrics import r2_score

r2_score(y_test,y_pred)


from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
reg.score(x_test,y_test)


from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 1000, random_state = 1,max_depth = 8, min_samples_split = 2,
          learning_rate = 0.1, loss = 'ls')

clf.fit(x_train, y_train)
clf.score(x_test,y_test)




from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score


from xgboost import XGBRegressor

np.random.seed(42)
rg7 = XGBRegressor()
rg7.fit(x_train, y_train)
y_pred2 = rg7.predict(x_test)

print(f"Score : {rg7.score(x_test, y_test)}")
print(f"Cross_val_score : {np.mean(cross_val_score(rg7, X, y, cv=5))}")
print(f"Root_mean_squared_error {np.sqrt(mean_squared_error(y_test, y_pred2))}")


r2_score2 = r2_score(y_test, y_pred2)
print(r2_score2)





from catboost import CatBoostRegressor

np.random.seed(42)
rg10 = CatBoostRegressor()
rg10.fit(x_train, y_train)
y_pred3 = rg10.predict(x_test)

print(f"Score : {rg10.score(x_test, y_test)}")
print(f"Cross_val_score : {np.mean(cross_val_score(rg10, X, y, cv=5))}")
print(f"Root_mean_squared_error {np.sqrt(mean_squared_error(y_test, y_pred3))}")


r2_score3 = r2_score(y_test, y_pred3)
print(r2_score3)








import streamlit as st
import numpy as np

st.set_page_config(page_title="My Webpage", page_icon=":tada:", layout="wide")
st.subheader("Hi i am  :wave:")
st.title("Data Scientist")
st.write("I am blue prismer and i love my lifde ")

st.title("Type of housing")
H_type = st.radio('Please select the type of your housing',
                     ['Appartment', 'House', 'Studio', 'Duplex'])


if H_type == 'Appartment':
    t = [1,0,0,0]
elif H_type == 'House':
    t = [0,1,0,0]
elif H_type == 'Studio':
    t = [0,0,1,0]
else:
    t = [0,0,0,1]

st.title("Neighborhood")
quartier_option = st.selectbox('Please select your Neighborhood',
                      ('','Quartier Chemins Des Vignes ','Quartier Colombiers ','Quartier Corentin Celton ',
                       'Quartier Epinettes ','Quartier Espaces Jeunes ','Quartier Forum Seine ','Quartier Jules Guesde Montesy ',
                       'Quartier Le Petit Pont Amphibie '))

#quartier_option = 'Quartier Epinettes '

Dict1 = {
 '': 0,  
 'Quartier Chemins Des Vignes ':1,
 'Quartier Colombiers ': 2,
 'Quartier Corentin Celton ': 3,
 'Quartier Epinettes ': 4,
 'Quartier Espaces Jeunes ': 5,
 'Quartier Forum Seine ': 6,
 'Quartier Jules Guesde Montesy ': 7,
 'Quartier Le Petit Pont Amphibie ':8 }

quartier_option = Dict1[quartier_option]




st.title("Energy Efficiency")
performance_e= st.select_slider('Please select the energy performance of your housing',
                     options=['','A','B','C','D','E','F'])

performance_e = 'A'

if performance_e == '':
    performance_e = 'Non Renseigné'
else:
    performance_e = performance_e.lower()
    


Dict2 = {
 'Non Renseigné': 0,  
 'a':1,
 'b': 2,
 'c': 3,
 'd': 4,
 'e': 5,
 'f': 6}

performance_e = Dict2[performance_e]


my_range = range(1,30)
st.title("Number of rooms")
nb_rooms = st.select_slider('Please select the number of rooms of your housing',
                     options=my_range, value =3)



my_range = range(1,30)
st.title("Number of bedrooms")
nb_bedrooms = st.select_slider('Please select the number of bedrooms of your housing',
                     options=my_range, value =2)




my_range = range(0,15)
st.title("Floor")
floor = st.select_slider('Please select the floor of your housing',
                     options=my_range, value =4)


my_range = range(9,600)
st.title("Living Square-feet")
sq_ft = st.select_slider('Please select the floor of your housing',
                     options=my_range, value =50)





st.title("Other house features")

 
cuisine_equipee = st.checkbox('Equipped kitchen')

if cuisine_equipee:

    cuisine_equipee = 1

else:

    cuisine_equipee = 0

 

cuisine_americaine  = st.checkbox('American style equipped kitchen')

if cuisine_americaine:

    cuisine_americaine = 1

else:

    cuisine_americaine = 0

 

ascenseur  = st.checkbox('Lift')

if ascenseur:

    ascenseur = 1

else:

    ascenseur = 0

 

digicode  = st.checkbox('Digicode')

if digicode:

    digicode = 1

else:

    digicode = 0

 

interphone  = st.checkbox('Interphone')

if interphone:

    interphone = 1

else:

    interphone = 0

 

placards  = st.checkbox('Storage room')

if placards:

    placards = 1

else:

    placards = 0

 

terasse  = st.checkbox('Terasse')

if terasse:

    terasse = 1

else:

    terasse = 0

 

balcon = st.checkbox('Balcony')

if balcon:

    balcon = 1

else:

    balcon = 0

 

baignoire  = st.checkbox('Bathtub')

if baignoire:

    baignoire = 1

else:

    baignoire = 0

 

sans_vis_a_vis = st.checkbox('Without Vis à Vis')

if sans_vis_a_vis:

    sans_vis_a_vis = 1

else:

    sans_vis_a_vis = 0

 

parking = st.checkbox('Available parking lot')

if parking:

    parking = 1

else:

    parking = 0

 

cave = st.checkbox('Basement')

if cave:

    cave = 1

else:

    cave = 0

 

jardin = st.checkbox('Backyard')

if jardin:

    jardin = 1

else:

    jardin = 0

 

gardien = st.checkbox('Builiding Castodian')

if gardien:

    gardien = 1

else:

    gardien = 0

 

CI = st.checkbox('Individual heating')

if CI:

    CI = 1

else:

    CI = 0

 

CIE = st.checkbox('Individual electric heating')

if CIE:

    CIE = 1

else:

    CIE = 0

 

CIG = st.checkbox('Individual gas heating')

if CIG:

    CIG = 1

else:

    CIG = 0

 

CCG = st.checkbox('Communal heating')

if CCG:

    CCG = 1

else:

    CCG = 0

 

expo_sud = st.checkbox('Southern exposure')

if expo_sud:

    expo_sud = 1

else:

    expo_sud = 0

 

 

#### /!\ A REMETTRE DANS LE BON ORDRE !!!!!!

x = np.array([nb_rooms, nb_bedrooms, floor, sq_ft,t[0],t[1],t[2],t[3],quartier_option,performance_e,
              cuisine_equipee, cuisine_americaine, ascenseur, digicode,
              interphone, placards, terasse, balcon, baignoire, sans_vis_a_vis,
              parking, cave, jardin, gardien, CI, CIE, CIG, CCG, expo_sud])


xx= x.reshape((1, 29))


y_pred_app = rg7.predict(xx)

st.write("The market price at which you can sell your house is : ",y_pred_app)

```
